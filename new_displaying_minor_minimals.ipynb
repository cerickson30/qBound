```python
def read_minor_minimals(nn, edge_count):
    import ast
    
    path_prefix = 'minor_minimal_graphs'
    
    minimals_list = []

    with open(path_prefix + f'/all_minimals/all_minimals_after_n_{nn}_edges-{edge_count}.txt', 'r') as infile:
        stored_minimals = infile.read()

        minimals_list.extend(ast.literal_eval(stored_minimals))
        
    return minimals_list
```

```python
nn = 9
edge_count = 1
minimals_list = read_minor_minimals(nn, edge_count)
```

```python
total = 0
for graph_list in minimals_list:
    total += len(graph_list)
    
print(f'The number of minor minimal graphs through {nn} vertices is: {total}')
```

```python
for ii in range(len(minimals_list)):
    vertices = len(Graph(minimals_list[ii][0]))
    print(f'Vertices: {vertices},\tMinor minimal graphs on {vertices} vertices: {len(minimals_list[ii])}')
```

```python
def display_minimal_graphs(nn, minimals_list):
    for g in minimals_list[nn]:
        Graph(g).show()
```

```python
for kk in range(1,5):
    
    print(f'Minor minimal graphs with spectator number = {kk}')
    display_minimal_graphs(kk-1, minimals_list)
    print(end='\n\n')
```
