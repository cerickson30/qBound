```python
import os
```

```python
os.listdir('data')
```

```python
for thing in os.walk('data'):
    print(len(thing))
    print(thing, end='\n\n')
```

```python
for path, list_of_folders, list_of_files in os.walk('data'):
    print(path)
#     print(list_of_files, end='\n\n')
    for filename in list_of_files:
        if 'backup' in filename:
            backup_file = filename
            os.remove(os.path.join(path, backup_file))
        else:
            print(os.path.join(path, filename))
```
