```python
# making complete minimals_dict

v6_path = 'data/minimals_dict'
v6_file_9_1 = 'minimals_dict_9_verts_1_edges.txt'

with open(os.path.join(v6_path, v6_file_9_1)) as infile:
    v6_minimals_dict_9_1 = eval(infile.read())
    
v6_file_10_1 = 'minimals_dict_10_verts_1_edges.txt'

with open(os.path.join(v6_path, v6_file_10_1)) as infile:
    v6_minimals_dict_10_1 = eval(infile.read())

    
    
minimals_dict = {}

for spec_key in v6_minimals_dict_9_1.keys():
    minimals_dict[spec_key] = v6_minimals_dict_9_1[spec_key].union(v6_minimals_dict_10_1[spec_key])
```

```python
minimals_dict.keys()
```

```python
for spec_key in minimals_dict.keys():
    for g6_str in minimals_dict[spec_key]:
        G = Graph(g6_str)
        kk = int(spec_key[0])
        if G.num_edges() >= 3*kk + 1:
            print(f'{g6_str}:\t {G.num_verts()} vertices,\t spectator floor = {kk},\t 3k+1 = {3*kk+1}, {G.num_edges()} edges')
        elif G.num_edges() >= 3*kk:
            print(f'{g6_str}:\t {G.num_verts()} vertices,\t spectator floor = {kk},\t 3k = {3*kk}, {G.num_edges()} edges')
```

```python
for spec_key in minimals_dict.keys():
    print(f'Number of minor-minimal connected graphs with spectator floor = {spec_key[0]}: {len(minimals_dict[spec_key])}')
```
